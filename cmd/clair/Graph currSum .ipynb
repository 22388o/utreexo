```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("result.csv")
```

```python
data
```

```python
data.columns = ["Block Number","Maximum number of remembers"]
data
```

```python
ax = data.plot.line(x="Block Number",y="Maximum number of remembers")
ax.set_title('100000 Blocks with 10 Look Ahead')
ax
```

```python
ax.figure.savefig('100000 Blocks with 10 Look Ahead.png')
```
