```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
all_data = pd.read_csv("resultAllThree.csv")
```

```python
all_data
```

```python
all_data.columns = ['Memory Size','Number of LookAhead Forgotten', 'Number of LookBehind Forgotten','Number of Clairvoy Forgotten']
all_data
```

```python
plt.plot(all_data['Memory Size'],all_data['Number of LookAhead Forgotten'], label = 'Number of LookAhead Forgotten')
plt.plot(all_data['Memory Size'],all_data['Number of LookBehind Forgotten'],label = 'Number of LookBehind Forgotten')
plt.plot(all_data['Memory Size'],all_data['Number of Clairvoy Forgotten'],label = 'Number of Clairvoy Forgotten')
plt.legend()
plt.show()
plt.savefig("All Three graphed for powers of 10.png")
```
