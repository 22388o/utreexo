```python
all_data = pd.read_csv("resultClear.csv")
all_data.columns = ['Reset Size','Number of LookAhead Forgotten', 'Number of LookBehind Forgotten','Number of Clairvoy Forgotten']
all_data
```

```python
new_data = pd.DataFrame()
for i,r in all_data.iterrows():
    new_data = new_data.append({'Reset Size':r['Reset Size'],'Number':r['Number of LookAhead Forgotten'],'Type':"Lookahead forgotten"},ignore_index=True)
    new_data = new_data.append({'Reset Size':r['Reset Size'],'Number':r['Number of LookBehind Forgotten'],'Type':"Lookbehind forgotten"},ignore_index=True)
    new_data = new_data.append({'Reset Size':r['Reset Size'],'Number':r['Number of Clairvoy Forgotten'],'Type':"Clairvoy forgotten"},ignore_index=True)
new_data
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="Reset Size", hue="Type", y="Number", data=new_data)
plt.savefig("All Three bar graphed for reset size.png")
```
